```python
# V3
###############
# [x[0] for x in os.walk("composite_jsons/")]
root = "composite_jsons/"
save_dir = "cleaner_labels/"
split_dir = "train_test_val_splits//"

key_list = ["gt_labels", "metadata"]

lift_phase_list = ["setup", "ascent", "top", "descent"]

lift_type_dict = {}
lift_type_dict["squat"] = ["setup", "descent", "bottom", "ascent"]
lift_type_dict["benchpress"] = ["setup", "descent", "bottom", "ascent"]
lift_type_dict["deadlift"] = ["setup", "ascent", "top", "descent", "floor break"]

metadata_list = ["extension", "filename", "folder"]

split_list = ["test.json", "train.json"]
split_file_list = []

for s in split_list:
    with open(split_dir + s) as f:
        split_json = json.load(f)
    for val in split_json:
        split_file_list.append(val['composite_json'])
        
# print(split_file_list)
split_file_list_set = set(split_file_list)

dir_list = next(os.walk(root))[1]
for d in dir_list:
    file_list = next(os.walk(root + d))[2]
    file_list_set = set(file_list)
    
    intersect_set = file_list_set.intersection(split_file_list_set)
    
#     print(len(split_file_list_set))
#     print(len(file_list_set))
#     print(len(intersect_set))
#     print()

    
    flaw_dict = {"data": [], "index": 0}
#     flaw_dict["data"] = []
#     flaw_dict["index"] = 0
    
    for file in intersect_set:
        with open(root + d + "/" + file) as f:
            comp_json = json.load(f)
        
        video_dict = {key_list[0]: {"lift_type": {}}, key_list[1]: {}}
#         video_dict[key_list[0]] = {}
#         video_dict[key_list[0]]["lift_type"] = {}
#         video_dict[key_list[1]] = {}
        
        for lp in lift_type_dict[d]:
            video_dict[key_list[0]]["lift_type"][lp] = []
        
        for m in metadata_list:
            video_dict[key_list[1]][m] = comp_json[key_list[1]][m]
            
        video_dict[key_list[1]]["macro_label"] = comp_json[key_list[0]]["lift_type"]["macro"]
        video_dict[key_list[1]]["micro_label"] = comp_json[key_list[0]]["lift_type"]["micro"]

        flaw_dict["data"].append(video_dict)
        
    save_file_path = save_dir + d + "_flaw.json"
    
    os.makedirs(os.path.dirname(save_file_path), exist_ok=True)
    with open(save_file_path, 'w') as f:
        json.dump(flaw_dict, f, indent=4, separators=(',', ': '), sort_keys=True)
```
